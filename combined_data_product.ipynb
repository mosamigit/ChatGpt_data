```python
new_cols = ['name', 'description', 'purl', 'manufacturer','cell','wattage_per_panel','weight','commercial_residential','min_wattage','cell_technology','dimension_or_size','breadth', 'thickness']
data = data[new_cols]
```

```python
data = data.astype(str)
print(data.dtypes)
```

```python
product_strings = []
for index, row in data.iterrows():
    product_name = row['name']
    product_url = row['purl']
    product_string = f'[{product_name}]("{product_url}")'
    product_strings.append(product_string)

# Add the product strings as a new column in the DataFrame
data['product_string'] = product_strings
```

```python
data2 = data['cell_dimension'] = data['dimension_or_size'] + ' x '+ data['breadth'] + ' x ' + data['thickness']+'mm'
```
