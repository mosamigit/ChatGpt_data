Data cleaning
1.Remove HTML tags from description column


```python
data['description'] = data['description'].str.replace(r'<[^<>]*>', '', regex=True)
```

```python
####Remove unnecessary string from the dataset
```

```python
data['description']= data['description'].str.split('Product Description').str[-1]
```

```python
data['description']
```

```python
data['description']= data['description'].str.replace('Inxeption Energy Solutions', '')
```

```python
data.head()
```

```python
data.columns
```

```python
new_cols = ['name', 'description', 'purl', 'manufacturer','cell','wattage_per_panel','weight','commercial_residential','min_wattage','cell_technology','dimension_or_size','breadth', 'thickness']
data = data[new_cols]
```

```python
data
```

```python
data.columns
```

```python
data.dtypes
```

```python
data = data.astype(str)
print(data.dtypes)
```

```python
product_strings = []
for index, row in data.iterrows():
    product_name = row['name']
    product_url = row['purl']
    product_string = f'[{product_name}]("{product_url}")'
    product_strings.append(product_string)

# Add the product strings as a new column in the DataFrame
data['product_string'] = product_strings
```

```python
data['product_string']
```

```python
data.dtypes
```

```python
data['cell_dimension'] = data['dimension_or_size'] + ' x '+ data['breadth'] + ' x ' + data['thickness']+'mm'
```

```python
data['cell_dimension']
```

```python
data['cell_dimension'] = data['cell_dimension'].apply(lambda x: x.replace('.0', '') if isinstance(x, str) else x)
```

```python
data['completion'] = "This is a "+ data['cell_technology'] + " solarpanel with a nominal wattage power of " +data['wattage_per_panel']+' watts.'+"The panel has "+data['cell']+"cells and measures "+data['cell_dimension'] +" in cell dimensions."+"This is manufacture by "+data['manufacturer']+"."+"The panel weighs "+data['weight']+"kg and suitable for "+data['commercial_residential']+" applications."+"key features:"+data['description']+"To know more "+data['product_string']
```
