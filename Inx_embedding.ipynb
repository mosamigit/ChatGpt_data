```python
import tiktoken 

tokenizer = tiktoken.get_encoding("cl100k_base")
df = pd.read_excel("31-60_products_new.xlsx", index_col = 0)
df.columns = ['name' , 'completion']

df['n_tokens'] = df.completion.apply(lambda x: len(tokenizer.encode(x)))

df.n_tokens.hist()
```
